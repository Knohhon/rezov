```python
import pandas as pd

vacancy_path = input()
vacancy = pd.read_csv(vacancy_path)
vacancy.dropna(axis=0, how='any', inplace=True)
res = vacancy.values.tolist()

def res_in_str(res):
	return [[str(x) if '.0' not in str(x) else str(x)[:-2] for x in line] for line in res]
	
print(list(vacancy.columns.values))
print(res_in_str(res))
```

```python
from prettytable import PrettyTable, ALL
import csv
import re
import os

table = PrettyTable()
tax = {'Да': 'Без вычета налогов', 'Нет': 'С вычетом налогов'}
exp = {"noExperience": "Нет опыта",
       "between1And3": "От 1 года до 3 лет",
       "between3And6": "От 3 до 6 лет",
       "moreThan6": "Более 6 лет"}
money = {"AZN": "Манаты",
         "BYR": "Белорусские рубли",
         "EUR": "Евро",
         "GEL": "Грузинский лари",
         "KGS": "Киргизский сом",
         "KZT": "Тенге",
         "RUR": "Рубли",
         "UAH": "Гривны",
         "USD": "Доллары",
         "UZS": "Узбекский сум"}
clean = re.compile('<[^>]+>')
about_salary = [0]*3
alf = {'name': 'Название',
       'description': 'Описание',
       'key_skills': 'Навыки',
       'experience_id': 'Опыт работы',
       'premium': 'Премиум-вакансия',
       'employer_name': 'Компания',
       'salary': 'Оклад',
       'area_name': 'Название региона',
       'published_at': 'Дата публикации вакансии'}


def rus_exp(text):
    for key in exp:
        if key in text:
            text = exp[key]
    return text


def date(text):
    normal_date = f"{text[8:10]}.{text[5:7]}.{text[:4]}"
    return normal_date


def normal_money(text):
    text = [i for i in text[::-1]]
    s = ''
    for x in range(len(text)):
        s += text[x]
        if (x + 1) % 3 == 0:
            s += ' '
    s = [j for j in s[::-1]]
    s = ''.join(s)
    return s.strip()

def skills(text):
    text = text.split(", ")
    out_text = "\n".join(text)
    return out_text


def len_text(text):
    if len(text) > 100:
        text = text[:100] + '...'
    return text


def print_vacancies(part_vacancy, i, row):
    if i < 7:
        if i == 3:
            row[i] = skills(part_vacancy)
        else:
        	row[i] = f"{part_vacancy.strip()}"
    elif 7 <= i < 10:
        about_salary[i-7] = part_vacancy
    elif i == 10:
        s1 = normal_money(str(int(float(about_salary[0]))))
        s2 = normal_money(str(int(float(about_salary[1]))))
        row[i - 3] = f"{s1} - {s2} ({money[part_vacancy]}) ({tax[about_salary[2]]})"
    elif i == 12:
        row[i - 3] = f"{date(part_vacancy)}"
    else:
        row[i - 3] = f"{part_vacancy.strip()}"
    if i - 2 == len(field):
        row = [len_text(w) for w in row]
        table.add_row(row)


def space_cleaner(text_with_excess_space):
    box = text_with_excess_space
    while any(chars in box for chars in ['  ', chr(160), chr(8194)]):
        if '  ' in box:
            box = box.replace('  ', ' ')
        elif chr(160) in box:
            box = box.replace(chr(160), ' ')
        else:
            box = box.replace(chr(8194), ' ')
    return box


def yesno(text):
    if text == 'True':
        return 'Да'
    if text == 'False':
        return 'Нет'


def cleaner(text):
    if '\n' in text:
        text = text.split('\n')
        text = [x.strip() for x in text]
        text = ', '.join(text)
    return re.sub(clean, '', text)


path = input()


if os.stat(path).st_size == 0:
    print("Пустой файл")
else:
    with open(path, encoding='utf-8-sig') as file:
        reader = csv.reader(file)
        res = [i for i in reader]
        if len(res) == 1:
            print("Нет данных")
        else:
            field = ['№'] + list(alf.values())
            table.field_names = field
            field_max = {key: 20 for key in field}
            table._max_width = field_max
            c = 0
            for line in res[1:]:
                if len(line) == 0:
                    print("Нет данных")
                if len(res[0]) == len(line) and '' not in line:
                    row = [''] * len(field)
                    row[0] = f"{c + 1}"
                    i = 1
                    for part in line:
                        part = cleaner(part)
                        part = space_cleaner(part)
                        part = rus_exp(part)
                        if i == 5 or i == 9:
                            part = yesno(part)
                        if i == 4:
                            part = rus_exp(part)
                        if part is None:
                            part = "Нет данных"
                        else:
                            print_vacancies(part, i, row)
                        i += 1
                    c += 1
                table.border = True
                table.align = "l"
                table.hrules=ALL
            print(table.get_string())
```
